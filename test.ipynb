```python
# Extract the data for plotting
x = results['driverId']
# Replace non-numeric values with NaN and drop them
height = pd.to_numeric(results['fastestLapSpeed'], errors='coerce').dropna()

# Ensure x and height have the same length after dropping NaN values
x = x[height.index]

# Create the bar plot
plt.bar(x=x, height=height)
plt.xlabel('Driver ID')
plt.ylabel('Fastest Lap Speed')
plt.title('Fastest Lap Speed by Driver')
plt.show()
```
