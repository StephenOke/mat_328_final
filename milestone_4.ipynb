Plot 1: NUMBER OF RACES PER YEAR

```python
races = races.dropna(subset=['date', 'year'])
races['date'] = races['date'].astype(str)
races['year'] = races['year'].astype(str)
races['date'] = pd.to_datetime(races['date'], errors='coerce')
races['year'] = pd.to_numeric(races['year'], errors='coerce')

last_10_years = races[(races['year'] >= 2014) & (races['year'] <= 2024)]
race_counts = last_10_years['year'].value_counts().sort_index()

# Plot the distribution of races per year
plt.bar(race_counts.index, race_counts.values, color='red', alpha=0.7)
plt.title("Distribution of F1 Races over the last 10 years")
plt.xlabel("Year")
plt.ylabel("Number of Races")

plt.show()
```

over the last 10 years, There had been a steadycount of between 18 to 22 races per year but covid 2020 had an effect F1 as there were on 16 races
there was a spike in 2024 due to the new found popularity in the sport with the likes of lewis Hamilton and Leclerc , with thier collaboration with streetwear brands so naturally there is an increase in amount of races held to feed the gained traction

**Distribution of races over months**

```python
if 'month' not in races.columns:
    races['month'] = pd.to_datetime(races['date'], errors='coerce').dt.month

races['month'].value_counts().sort_index().plot(kind='bar', color='orange')
plt.title('Distribution of F1 Races by Month')
plt.xlabel('Month (1 = Jan, 12 = Dec)')
plt.ylabel('Number of Races')
```

Most F1 races occur between March and November, with peaks in May, June, and October. There are almost no races during the winter months (December, January, February), which aligns with the sport’s off-season.
